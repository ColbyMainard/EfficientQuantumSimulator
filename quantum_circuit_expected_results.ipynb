# Expected Results - Various Quantum Circuits

```python
import cirq
import matplotlib.pyplot as plt
import pandas as pd
```

## Single Gate Circuits

```python
s=cirq.Simulator()

# X
x_circuit = cirq.Circuit()
a = cirq.LineQubit.range(1)[0]
x_circuit.append(cirq.X(a))
results=s.simulate(x_circuit)
x_circuit.append(cirq.measure(a, key='result'))
samples=s.run(x_circuit, repetitions=1000)
cirq.plot_state_histogram(samples, plt.subplot())
plt.show()
plt.close()

hist_dict = {}
samples.data["result"]
for element in samples.data["result"]:
    if element in hist_dict:
        hist_dict[element] = hist_dict[element] + 1
    else:
        hist_dict[element] = 1
print(hist_dict)
```

## Multi Gate Circuits

```python
# Bell state adapted from https://www.qmunity.tech/tutorials/introduction-to-cirq

bell_circuit = cirq.Circuit()
a, b = cirq.LineQubit.range(2)
bell_circuit.append(cirq.H(a))
bell_circuit.append(cirq.CNOT(a,b))

print('Simulate the circuit:')
results=s.simulate(bell_circuit)
print(results)

bell_circuit.append(cirq.measure(a, b, key='result'))

print('Sample the circuit:')
samples=s.run(bell_circuit, repetitions=1000)
cirq.plot_state_histogram(samples, plt.subplot())
plt.show()
plt.close()
print("Done!")

hist_dict = {}
samples.data["result"]
for element in samples.data["result"]:
    if element in hist_dict:
        hist_dict[element] = hist_dict[element] + 1
    else:
        hist_dict[element] = 1
print(hist_dict)
```

```python
random_circuit = cirq.Circuit()
a,b,c,d = cirq.LineQubit.range(4)
random_circuit.append(cirq.H(a))
random_circuit.append(cirq.H(b))
random_circuit.append(cirq.H(c))
random_circuit.append(cirq.H(d))
random_circuit.append(cirq.CNOT(a,b))
random_circuit.append(cirq.CNOT(c,d))
random_circuit.append(cirq.X(a))
random_circuit.append(cirq.Y(b))
random_circuit.append(cirq.Z(c))
random_circuit.append(cirq.rx(.7).on(d))
random_circuit.append(cirq.ry(.7).on(a))
random_circuit.append(cirq.rz(.7).on(b))
random_circuit.append(cirq.S(c))
random_circuit.append(cirq.T(d))
random_circuit.append(cirq.CZ(a, b))
random_circuit.append(cirq.CZ(c, d))
random_circuit.append(cirq.TOFFOLI(a, b, c))
random_circuit.append(cirq.FREDKIN(b, c, d))
results=s.simulate(random_circuit)
random_circuit.append(cirq.measure(a, b, c, d, key='result'))
samples=s.run(random_circuit, repetitions=1000)
cirq.plot_state_histogram(samples, plt.subplot())
plt.show()
plt.close()

hist_dict = {}
samples.data["result"]
for element in samples.data["result"]:
    if element in hist_dict:
        hist_dict[element] = hist_dict[element] + 1
    else:
        hist_dict[element] = 1
print(hist_dict)
```
